```python
useducation_df["AVG_MATH_4_SCORE"].fillna(useducation_df["AVG_MATH_4_SCORE"].mean(), inplace=True)
useducation_df["AVG_MATH_8_SCORE"].fillna(useducation_df["AVG_MATH_8_SCORE"].mean(), inplace=True)
useducation_df["AVG_READING_4_SCORE"].fillna(useducation_df["AVG_READING_4_SCORE"].mean(), inplace=True)
useducation_df["AVG_READING_8_SCORE"].fillna(useducation_df["AVG_READING_8_SCORE"].mean(), inplace=True)
```

```python
fill_list = ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE",
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE",
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G",
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G",
             "GRADES_9_12_G", "GRADES_ALL_G"]

states = useducation_df["STATE"].unique()

for state in states:
    useducation_df.loc[useducation_df["STATE"] == state, fill_list] = useducation_df.loc[useducation_df["STATE"] == state, fill_list].interpolate()
```

```python
# we drop the null values after interpolation
useducation_df.dropna(inplace=True)
```

```python

useducation_df["overall_score"] = (useducation_df["GRADES_4_G"]*((useducation_df["AVG_MATH_4_SCORE"] + useducation_df["AVG_READING_4_SCORE"])*0.5) + useducation_df["GRADES_8_G"]
                                 * ((useducation_df["AVG_MATH_8_SCORE"] + useducation_df["AVG_READING_8_SCORE"])*0.5))/(useducation_df["GRADES_4_G"] + useducation_df["GRADES_8_G"])
```

We weighted the score variables using the number of students in the respective grades.

```python
useducation_df[["overall_score", "TOTAL_EXPENDITURE", "INSTRUCTION_EXPENDITURE",
              "SUPPORT_SERVICES_EXPENDITURE", "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE"]].corr()
```

The most correlated variable with the overall score is the instruction expenditure

```python
X = useducation_df[["INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
                  "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE"]]

X = StandardScaler().fit_transform(X)

sklearn_pca = PCA(n_components=1)
useducation_df["pca_1"] = sklearn_pca.fit_transform(X)

print(
    'The percentage of total variance in the dataset explained by each',
    'component from Sklearn PCA.\n',
    sklearn_pca.explained_variance_ratio_
)
```


More than 94% of the total variance is explained by the first principal component.

```python
useducation_df[["overall_score", "pca_1", "TOTAL_EXPENDITURE", "INSTRUCTION_EXPENDITURE",
              "SUPPORT_SERVICES_EXPENDITURE", "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE"]].corr()
```

The correlation of the first principal component is less than the instruction expenditure and the support service expenditure variables.
